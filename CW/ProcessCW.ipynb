# Unzip and sort files

```python
import zipfile

def unzip_file(zip_src, dst_dir):
    r = zipfile.is_zipfile(zip_src)
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
    if r:     
        fz = zipfile.ZipFile(zip_src, 'r')
        for file in fz.namelist():
            fz.extract(file, dst_dir)    
    else:
        print('This is not zip')
        
PATH = "G:/_Code/coursework/unzip/"
```

```python
# create folders

for ID in df.IDs.unique():
    os.makedirs(PATH+ID)
```

```python
# move file and unzip file
import shutil
for i in range(len(df)):
    fname = df.iloc[i,0]
    ID = df.iloc[i,1]
    if df.iloc[i,2]!='zip':
        shutil.move(fname, PATH+ID)
    else:
        unzip_file(fname,PATH+ID)
```

```python
df_true = pd.read_csv('../test_results.csv') # true value
df_test = pd.read_csv('../Tweets_test.csv')
df_true['text'] = df_test['text']
df_true = df_true[['tweet_id','text','airline_sentiment']]
df_true.head()
```

```python
df_res = pd.DataFrame.merge(df_test, df_true, 'inner', on=['text'])
y_pred = df_res.airline_sentiment_x
y_true = df_res.airline_sentiment_y
```

```python
print(round(metrics.accuracy_score(y_true, y_pred),4),round(metrics.f1_score(y_true, y_pred,average='macro'),4),
     round(metrics.f1_score(y_true, y_pred,average='weighted'),4))
```

```python
# define metrics function
# df_true = pd.read_csv('../test_results.csv')# true value
def getMetrics(df_student, join_col='tweet_id'):
    df_res = pd.DataFrame.merge(df_student, df_true, 'inner', on=[join_col])
    y_pred = df_res.airline_sentiment_x
    y_true = df_res.airline_sentiment_y
    acc = round(metrics.accuracy_score(y_true, y_pred),4)
    f1_micro = round(metrics.f1_score(y_true, y_pred,average='micro'),4)
    f1_macro = round(metrics.f1_score(y_true, y_pred,average='macro'),4)    
    f1_weighted = round(metrics.f1_score(y_true, y_pred,average='weighted'),4)
    return acc, f1_micro, f1_macro, f1_weighted
```

```python
# check col names
cols_y, cols_x, csv_len, isYCorrect, isTweetID, isLenCorret = [],[],[],[],[],[]
acc, f1_micro, f1_macro, f1_weighted = [],[],[],[]
for fcsv in df_fcsv[df_fcsv['isCSV']>0].fcsv:
    fcsv = fcsv[0]
    df_student = pd.read_csv(fcsv,encoding = 'unicode_escape')
    cols_y.append(df_student.columns[-1]) # name of sentiment column
    df_student = df_student.rename(columns={cols_y[-1]:'airline_sentiment'}) # rename sentiment column
    if isinstance(df_student['airline_sentiment'][0], str): # if the sentiment label is string:
        df_student['airline_sentiment'] = df_student.airline_sentiment.apply(str).apply(str.lower) # lowercase sentiment labels
    
    cols_x.append(list(df_student.columns[:-1])) # name of tweet_id or text columns
    isTweetID.append(1 if 'tweet_id' in list(df_student.columns) else 0) # if tweet_id column exist?
    csv_len.append(len(df_student)) # length of prediction rows
    isLenCorret.append(1 if len(df_student)==1464 else 0) # if the length correct? 
    isYCorrect.append(1 if df_student.iloc[0,-1] in ['negative','neutral','positive']  else 0) # if the predicted label is correct?

    if (isTweetID[-1]==1) & (isLenCorret[-1]==1) & (isYCorrect[-1]==1): # for the csv with tweet_ID and has correct length, calculate the score
        _acc, _f1_micro, _f1_macro, _f1_weighted = getMetrics(df_student, join_col='tweet_id')
    elif (isTweetID[-1]==0) & (isLenCorret[-1]==1) & (isYCorrect[-1]==1):
        df_student = df_student.rename(columns={'Tweet':'text'})
        _acc, _f1_micro, _f1_macro, _f1_weighted = getMetrics(df_student, join_col='text')
    else:
        _acc, _f1_micro, _f1_macro, _f1_weighted = 0,0,0,0
    acc.append(_acc)
    f1_micro.append(_f1_micro)
    f1_macro.append(_f1_macro)
    f1_weighted.append(_f1_weighted)
```
